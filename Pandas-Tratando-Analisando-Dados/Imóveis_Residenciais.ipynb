## Imóveis Residenciais

```python
import pandas as pd
```

```python
dados = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Alura/Pandas/dados/aluguel.csv', sep=';')
dados.head(10)
```

```python
list(dados['Tipo'].drop_duplicates())
```

```python
residencial = ['Quitinete',
 'Casa',
 'Apartamento',
 'Casa de Condomínio',
 'Casa de Vila',]
```

```python
dados_residencial = dados[dados['Tipo'].isin(residencial)]
```

```python
dados_residencial.head(10)
```

```python
dados_residencial.reset_index(drop = True, inplace = True)
dados_residencial.head(10)
```

```python
print(dados.shape)
print(dados_residencial.shape)
```
