```python
data
```

```python
df1 = pd.DataFrame(data)
```

```python
df1
```

```python
index = ['Linha' + str(i) for i in range(3)]
index
```

```python
df1.index = index
df1
```

```python
columns = ['Coluna' + str(i) for i in range(3)]
columns
```

```python
df1.columns = columns
df1
```

```python
df1_2 = pd.DataFrame(data=data, columns=columns, index=index)
df1_2
```
