```python
import pandas as pd
```

```python
dados = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Alura/Pandas/extras/dados/aluguel_amostra.csv', sep=';')
dados.head(10)
```

```python
valores = dados['Valor m2']
```

```python
Q1 = valores.quantile(.25)
Q3 = valores.quantile(.75)
IIQ = Q3 - Q1
limite_inferior = Q1 - 1.5 * IIQ
limite_superior = Q3 + 1.5 * IIQ
print(Q1)
print(Q3)
print(IIQ)
print(limite_inferior)
print(limite_superior)
```
