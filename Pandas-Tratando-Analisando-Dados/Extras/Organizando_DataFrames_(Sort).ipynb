```python
import pandas as pd
```

```python
data = [[1, 2, 3], [7, 8, 9], [4, 5, 6]]
data
```

```python
index = list('321')
index
```

```python
df = pd.DataFrame(data, index, list('ZYX'))
```

```python
df
```

```python
df.sort_index(inplace=True)
df
```

```python
df.sort_index(axis=1, inplace=True)
df
```

```python
df.sort_values('X', inplace=True)
df
```

```python
df.sort_values('3', inplace=True, axis=1)
df
```

```python
df.sort_index(inplace=True)
df
```

```python
df.sort_values(['Y', 'X', 'Z'], inplace=True)
df
```
