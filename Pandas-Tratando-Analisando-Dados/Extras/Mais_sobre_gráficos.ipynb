```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('figure', figsize=(20, 10))
```

```python
dados = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Alura/Pandas/extras/dados/aluguel.csv', sep=';')
dados.head(10)
```

```python
area = plt.figure()
```

```python
g1 = area.add_subplot(2, 2, 1)
g2 = area.add_subplot(2, 2, 2)
g3 = area.add_subplot(2, 2, 3)
g4 = area.add_subplot(2, 2, 4)
```

```python
g1.scatter(dados['Valor'], dados['Area'])
g1.set_title('Valor x Área')

g2.hist(dados['Valor'])
g2.set_title('Histograma')

dados_g3 = dados['Valor'].sample(100).reset_index(drop=True)
g3.plot(dados_g3)
g3.set_title('Amostra (Valor)')

grupos = dados.groupby('Tipo')['Valor']
g4.bar(grupos.mean().index, grupos.mean().values)
g4.set_title('Valor Médio por Tipo')
```
