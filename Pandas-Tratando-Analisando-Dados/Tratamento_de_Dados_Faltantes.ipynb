### Limpeza dados nulos coluna 'Valor'

```python
dados[dados['Valor'].isnull()]
```

```python
A = dados.shape[0]
dados.dropna(subset=['Valor'], inplace=True)
B = dados.shape[0]
A - B
```

```python
dados[dados['Valor'].isnull()]
```

### Limpeza dos condomínios nulos apenas nos imóveis do tipo 'Apartamento'

```python
dados[dados['Condominio'].isnull()].shape[0]
```

```python
dados[(dados['Tipo'] == 'Apartamento') & (dados['Condominio'].isnull())]
```

```python
selecao = (dados['Tipo'] == 'Apartamento') & (dados['Condominio'].isnull())
A = dados.shape[0]
dados = dados[~selecao]
B = dados.shape[0]
A - B
```

```python
dados[dados['Condominio'].isnull()].shape[0]
```

### Substituição dos valores nulos do Condomínio e do IPTU restantes por 0

```python
# Substituir todos os valores nulos restantes no DataFrame por 0:
# dados.fillna(0, inplace = True)
```

```python
# Se quiser colocar valores diferentes para os valores nulos, de acordo
# com a coluna, ou substituir apenas os valores de colunas específicas:
dados.fillna({'Condominio': 0, 'IPTU': 0}, inplace = True)
```
